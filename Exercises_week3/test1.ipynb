```python
import cv2
from matplotlib import pyplot as plt

plt.gray()
```

```python
img = cv2.imread("nose_left.png")

img_left = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray_left = cv2.cvtColor(img_left, cv2.COLOR_RGB2GRAY)

img = cv2.imread("nose_right.png")

img_right = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray_right = cv2.cvtColor(img_right, cv2.COLOR_RGB2GRAY)

f, (ax_left, ax_right) = plt.subplots(1, 2, figsize=(10,10))
ax_left.imshow(gray_left)
ax_right.imshow(gray_right)
```

```python
def compare(template,img):
    sum=0
    delta=cv2.absdiff(template,img)
    h,w=delta.shape
    for i in range(h):
        for j in range(w):
            sum+=delta[i][j]
    return sum
```

```python
s=compare(gray_left,gray_right)
```

```python
img1 = cv2.imread("nose1.png")
gray1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
img2 = cv2.imread("nose2.png")
gray2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
img3 = cv2.imread("nose3.png")
gray3 = cv2.cvtColor(img3, cv2.COLOR_BGR2GRAY)

f, (ax_1, ax_2, ax_3) = plt.subplots(1, 3, figsize=(10,10))
ax_1.imshow(gray1)
ax_2.imshow(gray2)
ax_3.imshow(gray3)
```

```python
s1=compare(gray_left,gray1)
s2=compare(gray_left,gray2)
s3=compare(gray_left,gray3)
```

```python
span = cv2.imread("nose_span.png")
gray_span = cv2.cvtColor(span, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_span)
```

```python
def scan(template,img_span,step):
    h,w=img_span.shape
    min=255*h*h
    for i in range(w):
        s=compare(template,img_span[:,i:i+step])
        if s<min: min=s
    return min
```

```python
scan(gray_left,gray_span,7)
```
